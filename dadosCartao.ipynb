Quantidade de IDs não bate com a quantidade de linhas

Verificar se temos duplicatas, contas repetidas.

```python
id_counts = df['ID'].value_counts() 
id_counts.head()
```

```python
id_counts.value_counts()
```

MASCARA BOOLEANA

Método para filtrar um array, ou série, obedecendo uma condição

```python
dupe_mask = id_counts == 2
dupe_mask[0:5]
```

```python
id_counts.index[0:5]
```

```python
dupe_ids = id_counts.index[dupe_mask]
```

Alteramos a variável dupe_ids para uma lista

```python
dupe_ids = list(dupe_ids)
len(dupe_ids)
```

```python
dupe_ids[0:5]
```

Filtrando o Dataframe para visualizar as colunas dos 3 primeiros IDs duplicados

```python
df.loc[df['ID'].isin(dupe_ids[0:3]),:].head(10)
```

Criando uma matriz booleana com o mesmo tamanho do Dataframe inteiro, usando a condição "é igual a zero"

```python
df_zero_mask = df == 0
```

Criando uma série booleana que identifique cada linha em que todos  os elementos a partir  da segunda coluna sejam 0.

```python
feature_zero_mask = df_zero_mask.iloc[:,1:].all(axis=1)
```

```python
sum(feature_zero_mask)
```

Limpando o Dataframe, eliminando as linhas com zeros, exceto pelo ID

```python
df_clean_1 = df.loc[~feature_zero_mask,:].copy()
```
